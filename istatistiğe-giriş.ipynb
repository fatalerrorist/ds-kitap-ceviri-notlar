## Nüfus ve Örneklem

Verinin istatistiksel analizinde kullanılan verileri nüfusun içerisinden nüfusu temsil edecek şekilde alınan örneklerden sağlarız.

Verinin doğru analizi ancak ve ancak, genel nüfusu doğru temsil edecek örneklem ile sağlanır.

<b>Nüfus</b>, bir veri kümesindeki tüm öğeleri içerirken,

<b>Örneklem</b> ise nüfustan alınan bir veya daha fazla gözlemden oluşur.

- Örneklem, bir veri kümesinin tümünü temsil eder. 
- Örneklem, nüfusun(popülasyonun) bir alt kümesidir.
- Örneklem, nüfus içerisinden birden fazla kez seçilebilir.

İstatistiksel çıkarım söz konusu olduğunda, örneklem üzerinden alınan bilgi nüfus hakkında genellemeler yapılmasına olanak sağlar. Bu sebeple örneklem seçiminde nüfusu temsil edecek **parametrelerin** bilinmesi ve örneklem seçiminde bu **parametrelere** göre seçim yapılması önemlidir.

Avrupalı erkeklerin kilosunu incelememiz gereken senaryoda tüm avrupalıları ele alıp kilolarını ölçemeyiz. Bu durumda avrupalı erkeklerin kilosunu ölçmek için bir örneklem seçmemiz gerekir. Örneğin, 1000 avrupalı erkeğin kilosunu ölçebiliriz. Unutmamalıyız ki,  örneklem aynı zamanda nüfusun **parametrelerini**  aynı derecede temsil etmelidir. Örneğin, bölgelerin nüfus oranlarına göre örneklem seçmek gerekir. Örneğin, 1000 avrupalı erkeğin kilosunu ölçerken, 500'ü Almanya'dan, 250'si Fransa'dan, 250'si İtalya'dan olmalıdır. Bu şekilde örneklem seçmek, nüfusun her türlü **paremetresini** en yakın şekilde temsil edecektir.

**Parametre**, popülasyonun(nüfus) ve örneklemin sayısal değerlerine verilen genel bir addır. Popülasyon ile örneklem arasındaki parametre benzerliği standart sapma ve normal dağılım ile ölçülebilir.

<b> İstatistik, </b> bir örneklemdeki verileri kullanarak bir nüfus hakkında çıkarımlarda bulunmaktır. Örneğin, bir örneklemdeki verileri kullanarak bir nüfusun ortalama değerini tahmin edebiliriz. Aynı zamanda istatistik örnekleri şunlardır:

- Örneklemin ortalaması
- Örneklemin sapması
- Örneklemin korelasyon katsayısı
- Örneklemin regresyon eğrisi
- Örneklemin dağılımı

**Örneklem Dağılımı**, popülasyondan seçilen örneklemin parametrik özelliklerinin olasılık dağılımıdır. Örnek verirsek 100 insandan popülasyondan rastgele seçilen 10 kişinin kaç kilo olduklarının veya ne kadar uzunluğa sahip olduklarının olasılıksal dağılımı örneklem dağılımıdır.

Bu 10 kişinin her birinin kilosu genel olarak ele alındığında bize örneklem dağılımını sunmuş olur.

Peki ya **Olasılıksal dağılım** nedir?
 
Elimizde bir zar olduğunu düşünelim. Bu zarı 10 defa atıyoruz ve her gelen değeri not ediyoruz ve genel olarak bu değerlerin her birinin kaç defa geldiğini de yanına yazıyoruz. İşte bu değerlerin sıklığı bize olasılıksal dağılımı sunmuş olur. Örneğin 10 defa atılan zarda 5 değerinin 3 defa geldiğini düşünelim. Bu durumda 5 değerinin olasılıksal dağılımı 3/10 olur. Bu gibi dağılımları grafikte göstermek istediğimizde bu gibi bir grafik ortaya çıkar.

```python
import matplotlib.pyplot as plt
import numpy as np

# Değerler ve sayıları
values = [1, 2, 3, 4, 5, 6]
counts = [2, 1, 1, 1, 3, 2]  # Örnek veriler

# Grafik oluşturma
fig, ax = plt.subplots()
ax.plot(values, counts, linewidth=1, color='blue')
ax.fill_between(values, counts, color='blue', alpha=0.2, hatch='\\')

# 5 değerini vurgula
ax.plot([5, 5], [0, counts[4]], linewidth=2, color='red')

# Grafik ayarları
plt.xticks(values)
plt.xlabel('Zar Değeri')
plt.ylabel('Yoğunluk')
plt.title('10 Defa Atılan Zarda 5 Değerinin Yoğunluğu')

plt.show()
```
