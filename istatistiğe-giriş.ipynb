# Python ile istatistiğe giriş
### Gerçek hayat örnekleriyle

Her gün sonu belirsiz durumlarla karşı karşıya kalıyoruz ve eksik verilere dayanarak kararlar vermek zorunda kalıyoruz: “Otobüse yetişir miyim? Hangi hisse senedini almalıyım? Hangi adamla evlenmeliyim? Bu ilacı almalı mıyım? Çocuklarıma aşı yaptırmalı mıyım?” Bu sorulardan bazıları, çok fazla bilinmeyen değişken içerdikleri için istatistik alanının ötesindedir. ( örn. “Hangi kişiyle evlenmeliyim?”). Ancak birçok durumda, istatistikler verilen bilgilerden maksimum düzeyde bilgi çıkarmaya yardımcı olabilir ve bildiklerimizi ve bilmediklerimizi açıkça ortaya koyar. Örneğin, “Bu ilaç mide bulantısına neden olabilir” veya “Bu ilacı almazsan ölebilirsin” gibi belirsiz bir ifadeyi, “Bu ilacı kullanırken bin hastadan üçü mide bulantısı yaşıyor. ” veya “Bu ilacı almazsanız, ölme ihtimaliniz %95'tir.” şeklinde açıklayabiliriz. İstatistikler olmadan, verilerin yorumlanması hızla büyük ölçüde kusurlu hale gelebilir. Örneğin, "Alman Tank Problemi" olarak da bilinen 2. Dünya Savaşı sırasında üretilen tahmini Alman tankı sayısını ele alalım. Standart istihbarat verilerinden aylık üretilen Alman tankı sayısının tahmini 1550 idi; ancak, gözlemlenen tank sayısına dayalı istatistiksel tahmin 327 idi ve bu, 342 olan gerçek üretim sayısına çok yakındı (http://en.wikipedia.org/wiki/German_tank_problem).

Genel olarak istatistik bu konuları kapsar:
- Soruyu netleştirmek
- Soruyu cevaplayacak değişkeni ve değişkenin ölçüsünü belirlemek.
- Gerekli örneklem büyüklüğünü belirlemek.
- Değişkenliği tanımlamak.
- Tahmini parametreler hakkında nicel ifadeler kullanmak.
- Verilerinize dayanarak tahminlerde bulunmak.

## Nüfus ve Örneklem

Verinin istatistiksel analizinde kullanılan verileri nüfusun içerisinden nüfusu temsil edecek şekilde alınan örneklerden sağlarız.

Verinin doğru analizi ancak ve ancak, genel nüfusu doğru temsil edecek örneklem ile sağlanır.

<b>Nüfus</b>, bir veri kümesindeki tüm öğeleri içerirken,

<b>Örneklem</b> ise nüfustan alınan bir veya daha fazla gözlemden oluşur.

- Örneklem, bir veri kümesinin tümünü temsil eder. Örneklem, nüfusun bir alt kümesidir.
- Örneklem, nüfus içerisinden birden fazla kez seçilebilir.
