Her gün belirsiz durumlarla karşı karşıya kalıyoruz ve eksik verilere dayanarak kararlar vermek zorunda kalıyoruz: “Otobüse yetişir miyim? Hangi hisse senedini almalıyım? Hangi adamla evlenmeliyim? Bu ilacı almalı mıyım? Çocuklarıma aşı yaptırmalı mıyım?” Bu sorulardan bazıları, çok fazla bilinmeyen değişken içerdikleri için istatistik alanının ötesindedir. ( örn. “Hangi kişiyle evlenmeliyim?”). Ancak birçok durumda, istatistikler verilen bilgilerden maksimum düzeyde bilgi çıkarmaya yardımcı olabilir ve bildiklerimizi ve bilmediklerimizi açıkça ortaya koyar. Örneğin, “Bu ilaç mide bulantısına neden olabilir” veya “Bu ilacı almazsan ölebilirsin” gibi belirsiz bir ifadeyi, “Bu ilacı kullanırken bin hastadan üçü mide bulantısı yaşıyor. ” veya “Bu ilacı almazsanız, ölme ihtimaliniz %95'tir.” şeklinde açıklayabiliriz. İstatistikler olmadan, verilerin yorumlanması hızla büyük ölçüde kusurlu hale gelebilir. Örneğin, "Alman Tank Problemi" olarak da bilinen 2. Dünya Savaşı sırasında üretilen tahmini Alman tankı sayısını ele alalım. Standart istihbarat verilerinden aylık üretilen Alman tankı sayısının tahmini 1550 idi; ancak, gözlemlenen tank sayısına dayalı istatistiksel tahmin 327 idi ve bu, 342 olan gerçek üretim sayısına çok yakındı (http://en.wikipedia.org/wiki/German_tank_problem).

Genel olarak istatistik bu konuları kapsar:
- Soruyu netleştirmek
- Soruyu cevaplayacak değişkeni ve değişkenin ölçüsünü belirlemek.
- Gerekli örneklem büyüklüğünü belirlemek.
- Değişkenliği tanımlamak.
- Tahmini parametreler hakkında nicel ifadeler kullanmak.
- Verilerinize dayanarak tahminlerde bulunmak.

## Nüfus ve Örneklem

Verinin istatistiksel analizinde kullanılan verileri nüfusun içerisinden nüfusu temsil edecek şekilde alınan örneklerden sağlarız.

Verinin doğru analizi ancak ve ancak, genel nüfusu doğru temsil edecek örneklem ile sağlanır.

<b>Nüfus</b>, bir veri kümesindeki tüm öğeleri içerirken,

<b>Örneklem</b> ise nüfustan alınan bir veya daha fazla gözlemden oluşur.

- Örneklem, bir veri kümesinin tümünü temsil eder. 
- Örneklem, nüfusun(popülasyonun) bir alt kümesidir.
- Örneklem, nüfus içerisinden birden fazla kez seçilebilir.

İstatistiksel çıkarım söz konusu olduğunda, örneklem üzerinden alınan bilgi nüfus hakkında genellemeler yapılmasına olanak sağlar. Bu sebeple örneklem seçiminde nüfusu temsil edecek **parametrelerin** bilinmesi ve örneklem seçiminde bu **parametrelere** göre seçim yapılması önemlidir.

Avrupalı erkeklerin kilosunu incelememiz gereken senaryoda tüm avrupalıları ele alıp kilolarını ölçemeyiz. Bu durumda avrupalı erkeklerin kilosunu ölçmek için bir örneklem seçmemiz gerekir. Örneğin, 1000 avrupalı erkeğin kilosunu ölçebiliriz. Unutmamalıyız ki,  örneklem aynı zamanda nüfusun **parametrelerini**  aynı derecede temsil etmelidir. Örneğin, bölgelerin nüfus oranlarına göre örneklem seçmek gerekir. Örneğin, 1000 avrupalı erkeğin kilosunu ölçerken, 500'ü Almanya'dan, 250'si Fransa'dan, 250'si İtalya'dan olmalıdır. Bu şekilde örneklem seçmek, nüfusun her türlü **paremetresini** en yakın şekilde temsil edecektir.

**Parametre**, popülasyonun sayısal değerlerine verilen genel bir addır. Popülasyon ile örneklem arasındaki parametre benzerliği standart sapma ve normal dağılım ile ölçülebilir. Bu konulara daha sonra değineceğiz.
